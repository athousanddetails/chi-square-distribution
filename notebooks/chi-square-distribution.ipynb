## The Chi-Square Distribution

The Chi-Square distribution is a probability distribution that arises in statistical inference and hypothesis testing. It is commonly used to analyze categorical data and test the independence between two categorical variables. It's categorized as a Continuous Distribution.

The Chi-Square distribution is characterized by its degrees of freedom (df), which determine the shape of the distribution. The degrees of freedom depend on the specific problem or test being performed. The formula to calculate the Chi-Square statistic is as follows:

χ² = Σ((O - E)² / E)

Where:

* χ² is the Chi-Square statistic.
* Σ represents the sum of the calculations for all the categories or cells.
* O is the observed frequency for each category.
* E is the expected frequency for each category.

The Chi-Square can be used for two specific tests, the goodness-of-fit test and a test of independence of the data you want to test. 

Goodness-of-Fit Test is used to determine if a sample comes from a population with a specific distribution. It compares the observed frequencies (O) with the expected frequencies (E) to understand if there's a fit between them. 
Test of Independence is used to determine if there is any association between to category variables. It compares the observed frequencies (O) in contigency to a table of expected frequencies (E) under the assumption that there is independence between the two variables.

Ultimately, in both tests, the Chi-Square distribution is used to determine the critical value for rejecting the null hypothesis.

In this case we will be focusing on Goodness-of-Fit test.

### How to use the Chi-Square Distribution with Python to analyze data and probabilities

```python
# Packages importing

import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Chi-Square Goodness-of-fit Test

def dchisquare(x, df):
    results = chi2.pdf(x, df)
    return results

def cchisquare(x, df):
    results = chi2.cdf(x, df)
    return results

def qchisquare(x, df):
    results = chi2.ppf(x, df)
    return results
```

```python
# Using the functions and ploting

df = 55
x = np.arange(-4, 4, 0.001)

pdfvalues = dchisquare(x, df)

plt.plot(x, pdfvalues)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Chi-Square Density')
plt.show()
```

```python
df = 55
x = np.arange(-4, 4, 0.001)

cdfvalues = cchisquare(x, df)

plt.plot(x, cdfvalues)
plt.xlabel('x')
plt.ylabel('Probability')
plt.title('Chi-Square Density')
plt.show()
```
